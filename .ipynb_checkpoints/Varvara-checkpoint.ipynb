<h1>Предобработка данных</h1>

```python
data = pd.read_csv('./data.csv', low_memory=False)
df = pd.DataFrame(data)
df = df.drop(["fullVisitorID"], axis=1)
```

```python
df.head()
```

```python
df['transactions'] = df['transactions'].fillna(0)
goal = df["transactions"]
df = df.drop(["transactions"], axis=1)
```

```python
df.head()
```

```python
df.columns[df.isna().any()].tolist()
```

```python
print(df['bounces'].median(), df['pageviews'].median(), df['timeOnSite'].mean(), df['transactionRevenue'].mean())
```

```python
df['bounces'] = df['bounces'].fillna(df['bounces'].median())
df['pageviews'] = df['pageviews'].fillna(df['pageviews'].median())
df['timeOnSite'] = df['timeOnSite'].fillna(df['timeOnSite'].mean())
df['transactionRevenue'] = df['transactionRevenue'].fillna(df['transactionRevenue'].mean())

df.head()
```

```python
df['city'].value_counts()
```

```python
df['country'].value_counts()
```

```python
df['deviceCategory'].value_counts()
```

```python
df['browser'].value_counts()
```

```python
df['channelGrouping'].value_counts()
```

```python
df['source'].value_counts()
```

```python
df = df.drop("city", axis=1)
```

```python
df.head()
```

```python
df["source"] = df["source"].astype('category')
df["source"] = df["source"].cat.codes

df["channelGrouping"] = df["channelGrouping"].astype('category')
df["channelGrouping"] = df["channelGrouping"].cat.codes

df["browser"] = df["browser"].astype('category')
df["browser"] = df["browser"].cat.codes

df["deviceCategory"] = df["deviceCategory"].astype('category')
df["deviceCategory"] = df["deviceCategory"].cat.codes

df["country"] = df["country"].astype('category')
df["country"] = df["country"].cat.codes

df.head()
```

```python
train_data, val_data, train_y, val_y = train_test_split(df, goal, test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(train_data, train_y)

predicted = knn.predict(val_data)
predicted

np.array(val_y)
accuracy_score(predicted, val_y)
```
